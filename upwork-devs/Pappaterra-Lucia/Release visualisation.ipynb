We get the json file with the information from
https://wmwaredata.s3.us-east-2.amazonaws.com/releases.json

```python
url = 'https://wmwaredata.s3.us-east-2.amazonaws.com/releases.json'
r = requests.get(url, allow_redirects=True)
open('data/releases.json', 'wb').write(r.content)
```

```python
df0 = pd.read_json('data/releases.json')
df0

COLUMN_NAMES = ['assets', 'assets_url', 'author', 'body', 'created_at', 'draft', 'html_url', 
                'id', 'name', 'node_id', 'prerelease', 'published_at', 'tag_name', 
                'tarball_url', 'target_commitish', 'upload_url', 'url', 'zipball_url']
df = pd.DataFrame(columns=COLUMN_NAMES)

for i in range(len(df0)):
    for j in range(len(df0.columns)):
        my_dict = df0.iloc[i][j]
        if my_dict is not None:
            df = pd.concat([df, pd.DataFrame([my_dict])])
df = df.reset_index(drop=True)

repos = []
for i in range(len(df)):
    s = df['assets_url'].iloc[i]
    repo = re.search('https://api.github.com/repos/k8-proxy/(.*)/releases/', s)
    repo = repo.group(1)
    repos.append(repo)
df["repo"] = repos

# Reorder columns
cols = list(df.columns)
cols = [cols[-1]] + cols[0:-1]
df = df[cols]

df
```

## Repo's Releases

```python
# CREATE JSON FILE WITH NODES AND EDGES

network = Network()

for repo in set(list(df['repo'])):
    network.add_node(label=repo, color=rgb2hex(plt.cm.Pastel1(0)), shape='box')
    df1 = df[df['repo'] == repo].reset_index()
    
    j = 0
    for i in range(len(df1)):
        label = 'Tag:' + df1['tag_name'].iloc[i] + '\nid:' + str(df1['id'].iloc[i])
        if label in network.get_labels(): j += 1
        network.add_node(label=label, color=rgb2hex(plt.cm.Pastel1(1)), repeat_nodes=True)
        network.add_edge(network.nodesIds[repo][0], network.nodesIds[label][j])
    
network.save_to_json('data/releases_nodes_edges_1.json')
```

## Repo's Releases with Description

```python
# CREATE JSON FILE WITH NODES AND EDGES

network = Network()
network.groups = ['Repo', 'Description']
network.groupSettings = group_settings_R

for repo in set(list(df['repo'])):
    network.add_node(label=repo, group='Repo')
    df1 = df[df['repo'] == repo].reset_index()
    
    j, k = 0, 0
    for i in range(len(df1)): 
        
        label = 'Tag:' + df1['tag_name'].iloc[i] + '\nid:' + str(df1['id'].iloc[i])
        if label in network.get_labels(): j += 1
        network.add_node(label=label, group='Tag_Id', repeat_nodes=True)
        network.add_edge(network.nodesIds[repo][0], network.nodesIds[label][j])
        
        body = wrap_by_word(re.sub('<.*?>', '', df1['body'].iloc[i]), n=5)
        body = wrap_by_char(body, n=45)
        if body in network.get_labels(): k += 1
        network.add_node(label=body, group='Description', repeat_nodes=True)
        network.add_edge(network.nodesIds[label][j], network.nodesIds[body][k], length=200)
    
network.save_to_json('data/releases_nodes_edges_2.json')
```
